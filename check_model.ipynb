## 作成したモデルが適切に予測しているかチェック

```python
import numpy as np

from load_data import load, load2d
from saver import load_arch, load_history
from plotter import plot_samples, plot_hist
import pickle
```

```python
# テストデータ読み込み
X_test, _ = load(test=True)
X_test2d, _ = load2d(test=True)
```

```python
# モデルの取得 & 予測
model1 = load_arch('model/model1-arch-100.json')
model1.load_weights('model/model1-weights-100.hdf5')
model1_hist = load_history('model1')
y_model1 = model1.predict(X_test)

model2 = load_arch('model/model2-arch-1000.json')
model2.load_weights('model/model2-weights-1000.hdf5')
model2_hist = load_history('model2')
y_model2 = model2.predict(X_test2d)

model3 = load_arch('model/model3-arch-3000.json')
model3.load_weights('model/model3-weights-3000.hdf5')
model3_hist = load_history('model3')
y_model3 = model3.predict(X_test2d)
```

```python
model4 = load_arch('model/model4-arch-1000.json')
model4.load_weights('model/model4-weights-1000.hdf5')
model4_hist = load_history('model4')
y_model4 = model4.predict(X_test2d)

model5 = load_arch('model/model5-arch-1000.json')
model5.load_weights('model/model5-weights-1000.hdf5')
model5_hist = load_history('model5')
y_model5 = model5.predict(X_test2d)

model6 = load_arch('model/model6-arch-5000.json')
model6.load_weights('model/model6-weights-5000.hdf5')
model6_hist = load_history('model6')
y_model6 = model6.predict(X_test2d)

model7 = load_arch('model/model7-arch-5000.json')
model7.load_weights('model/model7-weights-5000.hdf5')
model7_hist = load_history('model7')
y_model7 = model6.predict(X_test2d)
```

```python
# 各モデルの損失関数を表示
print('loss min:', np.min(model1_hist['loss']), 'val loss min:', np.min(model1_hist['val_loss']))
print('loss min:', np.min(model2_hist['loss']), 'val loss min:', np.min(model2_hist['val_loss']))
print('loss min:', np.min(model3_hist['loss']), 'val loss min:', np.min(model3_hist['val_loss']))
print('loss min:', np.min(model4_hist['loss']), 'val loss min:', np.min(model4_hist['val_loss']))
print('loss min:', np.min(model5_hist['loss']), 'val loss min:', np.min(model5_hist['val_loss']))
print('loss min:', np.min(model6_hist['loss']), 'val loss min:', np.min(model6_hist['val_loss']))
print('loss min:', np.min(model7_hist['loss']), 'val loss min:', np.min(model7_hist['val_loss']))
```
